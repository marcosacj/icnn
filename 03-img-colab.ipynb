# Images Load

```
import numpy as np
import cv2
import os
```

```
instances = []

for fn in os.listdir('test/') [:10]:
    
    fp = 'test/{0}'.format(fn)
    
    img = cv2.imread(fp, 1)
    
    instances.append(img)
```
